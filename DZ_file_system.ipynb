```python
import os
import shutil

def create_file(file_path):
    try:
        with open(file_path, 'w'):
            print(f"Файл {file_path} успешно создан")
    except Exception as e:
        print(f"Произошла ошибка при создании файла: {e}")

def delete_file(file_path):
    try:
        os.remove(file_path)
        print(f"Файл {file_path} успешно удален")
    except Exception as e:
        print(f"Произошла ошибка при удалении файла: {e}")

def create_directory(directory_path):
    try:
        os.mkdir(directory_path)
        print(f"Папка {directory_path} успешно создана")
    except Exception as e:
        print(f"Произошла ошибка при создании папки: {e}")

def delete_directory(directory_path):
    try:
        os.rmdir(directory_path)
        print(f"Папка {directory_path} успешно удалена")
    except Exception as e:
        print(f"Произошла ошибка при удалении папки: {e}")

def delete_Fulldirectory(directory_path):
    try:
        # Удаление папки
        shutil.rmtree(directory_path)
        print(f"Папка {directory_path} успешно удалена")
    except Exception as e:
        print(f"Ошибка при удалении папки: {e}")
        
def move_file(src, dest):
    try:
        shutil.move(src, dest)
        print(f"Файл успешно перемещен из {src} в {dest}")
    except Exception as e:
        print(f"Произошла ошибка при перемещении файла: {e}")

def copy_file(src, dest):
    try:
        shutil.copy(src, dest)
        print(f"Файл успешно скопирован из {src} в {dest}")
    except Exception as e:
        print(f"Произошла ошибка при копировании файла: {e}")

def main_menu():
    print("Добро пожаловать в консольный файловый проводник!")
    while True:
        print("\nМеню:")
        print("1. Создать файл")
        print("2. Удалить файл")
        print("3. Создать папку")
        print("4. Удалить пустую папку")
        print("5. Удалить папку с файлами")
        print("6. Переместить файл")
        print("7. Скопировать файл")
        print("8. Выйти")
        choice = input("Выберите операцию: ")
        
        if choice == "1":
            file_path = input("Укажите путь и имя нового файла: ")
            create_file(file_path)
        elif choice == "2":
            file_path = input("Укажите путь к файлу для удаления: ")
            delete_file(file_path)
        elif choice == "3":
            directory_path = input("Укажите путь и имя новой папки: ")
            create_directory(directory_path)
        elif choice == "4":
            directory_path = input("Укажите путь к папке для удаления: ")
            delete_directory(directory_path)
            
        elif choice == "5":
            directory_path = input("Укажите путь к папке для удаления: ")
            delete_Fulldirectory(directory_path)    
            
            
        elif choice == "6":
            src = input("Укажите путь к исходному файлу: ")
            dest = input("Укажите путь к папке куда переместить файл: ")
            move_file(src, dest)
        elif choice == "7":
            src = input("Укажите путь к исходному файлу: ")
            dest = input("Укажите путь к папке, куда скопировать файл: ")
            copy_file(src, dest)
        elif choice == "8":
            print("До свидания")
            break
        else:
            print("Неверный выбор. Попробуйте снова.")

if __name__ == "__main__":
    main_menu()
```
